```python
import yfinance as yf
import pandas as pd
import numpy as np

btc_ticker = "BTC-USD"
btc_data = yf.download(btc_ticker, start="2014-01-01", end="2024-12-31", interval="1d")
btc_data.reset_index(inplace=True)
print(btc_data.head())

btc_data.to_csv("bitcoin_10_years.csv", index=False)
print("Data saved to bitcoin_10_years.csv")

print("Missing values:\n", btc_data.isnull().sum())
print("Duplicate rows:", btc_data.duplicated().sum())

btc_data['Date'] = pd.to_datetime(btc_data['Date'])
```

```python
eth_ticker = "ETH-USD"
eth_data = yf.download(eth_ticker, start="2014-01-01", end="2024-12-31", interval="1d")
eth_data.reset_index(inplace=True)
print(eth_data.head())

eth_data.to_csv("ethereum_10_years.csv", index=False)
print("Data saved to ethereum_10_years.csv")

print("Missing values in Ethereum data:\n", eth_data.isnull().sum())
print("Duplicate rows:", eth_data.duplicated().sum())

eth_data['Date'] = pd.to_datetime(eth_data['Date'])
```

```python
btc_data = pd.read_csv("bitcoin_10_years.csv")
eth_data = pd.read_csv("ethereum_10_years.csv")
btc_data['Date'] = pd.to_datetime(btc_data['Date'])
eth_data['Date'] = pd.to_datetime(eth_data['Date'])
btc_data.rename(columns={
    'Open': 'BTC_Open',
    'High': 'BTC_High',
    'Low': 'BTC_Low',
    'Close': 'BTC_Close',
    'Adj Close': 'BTC_Adj_Close',
    'Volume': 'BTC_Volume'
}, inplace=True)

eth_data.rename(columns={
    'Open': 'ETH_Open',
    'High': 'ETH_High',
    'Low': 'ETH_Low',
    'Close': 'ETH_Close',
    'Adj Close': 'ETH_Adj_Close',
    'Volume': 'ETH_Volume'
}, inplace=True)
crypto_data = pd.merge(btc_data, eth_data, on='Date', how='inner')
crypto_data.set_index('Date', inplace=True)
print(crypto_data.head())
crypto_data.to_csv("crypto_merged.csv")
print("Merged dataset saved as crypto_merged.csv")
```

```python
crypto_data = pd.read_csv("crypto_merged.csv")
crypto_data['Date'] = pd.to_datetime(crypto_data['Date'], errors='coerce')
print("Dataset Info:")
print(crypto_data.info())
print("\nFirst 5 rows of the dataset:")
print(crypto_data.head())
print("\nSummary Statistics:")
print(crypto_data.describe())
print("\nDate Range:")
print(f"From {crypto_data['Date'].min()} to {crypto_data['Date'].max()}")
print("\nMissing Values:")
print(crypto_data.isnull().sum())
```

```python
crypto_data[['BTC_Close', 'ETH_Close']] = crypto_data[['BTC_Close', 'ETH_Close']].apply(pd.to_numeric, errors='coerce')
crypto_data.dropna(subset=['BTC_Close', 'ETH_Close'], inplace=True)
btc_first = crypto_data['BTC_Close'].iloc[0]
eth_first = crypto_data['ETH_Close'].iloc[0]
crypto_data['BTC_Normalized'] = crypto_data['BTC_Close'] / btc_first
crypto_data['ETH_Normalized'] = crypto_data['ETH_Close'] / eth_first

print(crypto_data[['Date', 'BTC_Close', 'ETH_Close', 'BTC_Normalized', 'ETH_Normalized']].head())
```

```python
crypto_data['BTC_Return'] = np.log(crypto_data['BTC_Close'] / crypto_data['BTC_Close'].shift(1))
crypto_data['ETH_Return'] = np.log(crypto_data['ETH_Close'] / crypto_data['ETH_Close'].shift(1))
crypto_data['BTC_Volatility'] = crypto_data['BTC_Return'].rolling(window=30).std()
crypto_data['ETH_Volatility'] = crypto_data['ETH_Return'].rolling(window=30).std()
crypto_data.dropna(inplace=True)
print(crypto_data[['Date', 'BTC_Close', 'ETH_Close', 'BTC_Return', 'ETH_Return', 'BTC_Volatility', 'ETH_Volatility']].head())
```

```python
crypto_data[['BTC_Close', 'ETH_Close']] = crypto_data[['BTC_Close', 'ETH_Close']].apply(pd.to_numeric, errors='coerce')
crypto_data.dropna(subset=['BTC_Close', 'ETH_Close'], inplace=True)
if 'BTC_Return' not in crypto_data.columns:
    crypto_data['BTC_Return'] = crypto_data['BTC_Close'].pct_change()
    crypto_data['ETH_Return'] = crypto_data['ETH_Close'].pct_change()

if 'BTC_Volatility' not in crypto_data.columns:
    crypto_data['BTC_Volatility'] = crypto_data['BTC_Return'].rolling(window=30).std()
    crypto_data['ETH_Volatility'] = crypto_data['ETH_Return'].rolling(window=30).std()

if 'BTC_Normalized' not in crypto_data.columns:
    btc_first = crypto_data['BTC_Close'].iloc[0]
    eth_first = crypto_data['ETH_Close'].iloc[0]
    crypto_data['BTC_Normalized'] = crypto_data['BTC_Close'] / btc_first
    crypto_data['ETH_Normalized'] = crypto_data['ETH_Close'] / eth_first
crypto_data.dropna(subset=['BTC_Return', 'ETH_Return', 'BTC_Volatility', 'ETH_Volatility'], inplace=True)
crypto_data.to_csv("crypto_ready.csv", index=False)

print("✅ Dataset successfully updated with all required columns.")
```
